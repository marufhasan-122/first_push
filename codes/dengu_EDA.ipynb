```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
df=pd.read_csv('dataset.csv')
print (df.head())
print (df.columns)
(df['Outcome'] == ((df['NS1'] == 1) | (df['IgG'] == 1) | (df['IgM'] == 1)).astype(int)).value_counts()
```

In the dataset NS1 means Non structural protein 1 Indicates:Acute infection , IgM means Immunoglobulin M (first antibody produced by the immune system when a virus attacks.) Indicates:Recent infection , IgG (Immunoglobulin G)An antibody that appears later in the immune response . Indicates:Past or secondary infection
In medical datasets like this, Outcome usually represents the final diagnosis result or disease confirmation. Based on your dataset:
Outcome = 1 probably means the patient tested positive for dengue.
Outcome = 0 means the patient tested negative for dengue.
So it's likely that:
If any of NS1, IgM, or IgG is positive, the Outcome should be 1.
If all are negative, the Outcome should be 0.
